```python
import pandas as pd
import numpy as np

from prediction_engine.data_fetch.fetch_data import fetch_data
from prediction_engine.pipeline_run.run_pipeline import run_prediction_pipeline
```

```python
print("Fetching data...")

df = fetch_data("RELIANCE.NS")

df = df.reset_index()

df["Date"] = pd.to_datetime(df["Date"])

df = df.sort_values("Date").reset_index(drop=True)

df.tail()
```

```python
test_df = df.iloc[-120:-10].copy()   # safe range

dates = test_df["Date"].sample(50, random_state=42).sort_values()

results = []
```

```python
for d in dates:

    try:
        out = run_prediction_pipeline(
            ticker="RELIANCE.NS",
            specified_date=str(d.date()),
            plot=False
        )

        pred_return = out["predicted_return"]

        p1 = df.loc[df["Date"] == d, "Close"].values[0]

        p2 = df.loc[df["Date"] > d, "Close"].iloc[0]

        real_return = (p2 - p1) / p1

        results.append({
            "date": d,
            "pred": pred_return,
            "real": real_return
        })

    except Exception as e:
        print("Skipped:", d.date())
```

```python
res = pd.DataFrame(results)

res.head()
```

```python
mae = np.mean(np.abs(res["pred"] - res["real"]))

direction_acc = np.mean(
    np.sign(res["pred"]) == np.sign(res["real"])
)

print("MAE =", mae)
print("Direction Accuracy =", direction_acc)
```
